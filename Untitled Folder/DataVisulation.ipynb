```r
choose_color = function(dsp,hwy){
    if ( dsp >= 5 & hwy >= 20 ) { return("red"); 
       } else {
        return("black");
    }
}
mycolor = NA
c = 1
while ( c <= length(mpg$hwy) ) { 
    mycolor[c] = choose_color(mpg$displ[c],mpg$hwy[c]);
    c = c + 1 ; 
    }
head(mycolor)
length(mycolor)
```

```r
ggplot(mpg)+geom_point(mapping=aes(x=displ,y=hwy),color=mycolor)
```
